# Assignment


### 1. Write a Python program to construct the following pattern, using a nested for loop.

```python
n=10
for i in range(0,n):
    for j in range(0,i+1):
        print("x ",end="")
    print("\r")                
```

### 2. Write a Python program to count the number of even and odd numbers from a series of numbers.

```python
number=[1,2,3,4,5,6,7,8,9]
even = [i for i in number if i%2 !=0]
odd = [i for i in number if i%2 ==0]
print("Number of even numbers :"+str(len(even)))
print("Number of odd numbers :"+str(len(odd)))
```
